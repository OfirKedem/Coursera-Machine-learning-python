# 2 Neural Networks

```python
import numpy as np
from scipy.io import loadmat
```

```python
data = loadmat('../data/ex3data1.mat')
y = data['y']
y %= 10 # fixes “0” digit is labeled as “10”
y = np.squeeze(y) # shape [N,1] -> [N,]
X = data['X'] 
X = np.append(np.ones([X.shape[0],1]),X, axis=1) # add bais
```

## 2.1 Model representation

```python
weights = loadmat('../data/ex3weights.mat')
Theta1 = weights['Theta1']
Theta2 = weights['Theta2']
print(Theta1.shape)
print(Theta2.shape)
```

## 2.2 Feedforward Propagation and Prediction

```python
def sig(x):
    return 1/(1+np.exp(-x))
```

```python
# first layer + activation
term1 = sig(X.dot(Theta1.T))
# add bais
term2 = np.append(np.ones([term1.shape[0],1]),term1, axis=1)
# second layer + activation
prob = sig(term2.dot(Theta2.T))
```

```python
pred = (prob.argmax(axis=1)+1)%10
acc = ((pred-y)==0).sum()/len(y)
print('The accuracy is: ' + str(acc))
```
