# 2 Spam Classification

```python
import numpy as np
from scipy.io import loadmat
from sklearn import svm
```

```python
data = loadmat('../data/spamTrain.mat', squeeze_me=True)
X = data['X']*1.0 # from dtype uint8 to float64
y = data['y']* -1 * -1 # uint8 -> int16

test_data = loadmat('../data/spamTest.mat', squeeze_me=True)
Xtest = test_data['Xtest']*1.0
ytest = test_data['ytest']* -1 * -1 
```

```python
clf = svm.SVC() # C=1.0, kernel='rbf', gamma='scale'
clf.fit(X,y)
```

```python
train_acc = 1 - np.abs(clf.predict(X) - y).sum()/len(y)
test_acc = 1 - np.abs(clf.predict(Xtest) - ytest).sum()/len(ytest)
```

```python
print('Train accuracy: %.3f \nTest accuracy: %.3f' % (train_acc, test_acc))
```
